```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('heart.csv', sep=',')
dataset
```

```python
x = np.asarray(dataset.drop(['target'], axis = 1), dtype = np.float64)
y = np.asarray(dataset['target'], dtype = np.int64)
print(x, y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
```

```python
heart = np.asarray([[63,0,3,145,233,1,0,150,0,2.3,0,0,1]])
print(model.predict(heart))
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(x_train)
print(accuracy_score(y_train, y_pred))

y_pred_test = model.predict(x_test)
print(accuracy_score(y_test,y_pred_test))
```
