```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('sales.csv',sep=',')
dataset
```

```python
data = np.asarray(dataset[['W0','W1']])
print(data)
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
```

```python
plt.title("Dendogram")
Dendrogram = sch.dendrogram(sch.linkage(data, method = 'ward'))
```

```python
plt.title("Dendogram")
Dendrogram = sch.dendrogram(sch.linkage(data, method = 'average'))
```

```python
plt.title("Dendogram")
Dendrogram = sch.dendrogram(sch.linkage(data, method = 'complete'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=5)
output = ac.fit_predict(data)

plt.scatter(data[:,0], data[:,1], s=50, c=output, cmap='rainbow')
plt.show()
```

```python
ac = AgglomerativeClustering(n_clusters=5, affinity='manhattan', linkage='average')
output = ac.fit_predict(data)

plt.scatter(data[:,0], data[:,1], s=50, c=output, cmap='rainbow')
plt.show()
```
